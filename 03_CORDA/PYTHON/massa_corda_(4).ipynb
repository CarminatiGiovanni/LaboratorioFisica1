```python
import numpy as np
from matplotlib import pyplot as plt
from interpolazione3 import decimal_val

f_FINE = np.array([89,183,275])
f_GROSSA = np.array([27,54,81])

massa_GROSSA = 4.3 * 0.001 #kg
massa_FINE = 0.3 * 0.001 # kg

l_GROSSA = 2 # m
l_FINE = 1.23 # m

mu_GROSSA = massa_GROSSA/l_GROSSA
mu_FINE = massa_FINE/l_FINE

n = np.array([1,2,3])

plt.plot(n,f_FINE,'o-', label=f'$\mu_1: {decimal_val(mu_GROSSA,exp=-4)}$')
plt.plot(n,f_GROSSA,'o-', label=f'$\mu_2: {decimal_val(mu_FINE,exp=-4)}$')
plt.xlabel('$N_{armonica}$')
plt.ylabel(f"${chr(92)}nu (Hz)$")
plt.xticks(n)
plt.legend()
plt.show()
```
